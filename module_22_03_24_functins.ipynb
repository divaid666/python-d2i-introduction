```python
def minus_xy(x,y):
    z = x - y  # this z variable exists only in the function - that variable has got function-level scope
    print(z)

minus_xy(10,7)
```

```python
# to make sure that the variable in the function has got global scope - use return
def minus_dt(d,t):
    z = d - t
    return z

answer = minus_dt(7,3)
print(answer)   # if there was no return z, then the print(answer) code would return none
```

```python
def maths_xy(x,y):
    minus = x - y
    plus = x + y
    return minus, plus

print(maths_xy(100,20)) 

print(maths_xy(100,20)[0])

x_minus_y, x_plus_y = maths_xy(100,20)
print(x_minus_y)
print(x_plus_y)
```

```python
def more_maths(x,y):
    times = x*y
    divide = x / y
    return times, divide

times_result, divide_result = more_maths(100,20)
print(times_result)
print(divide_result)
```

```python
def function_1(x,y):
    return f'function 1 says x and y are {x}, {y}'

def function_2 (x,y):
    return f'function 2 says x and y are {x}, {y}'

def nested_funct(j,k):
    var_1 = function_1(j,k)
    var_2 = function_2(j,k)
    return var_1, var_2

print(nested_funct('Hi','Hello'))
```

```python
def my_forename(forename):
    return f'Hello my name is {forename}'

def my_surname(surname):
    return f'My surname is {surname}'

def full_name_funct (fname, sname):
    return my_forename(fname), my_surname(sname)

names = full_name_funct(fname = "Dawid", sname="Tycki")
print(names)
```

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/1980%202023%20average%20house%20prices.csv"
```

```python
df = pd.read_csv(url)
df
```

```python
mean_pct_change = df['Percentage change (monthly) All property types'].mean()
print(mean_pct_change)
```

```python
mean_pct_change = df['Percentage change (monthly) All property types'].mean()

def change_checker(pct_change):
    if pct_change > mean_pct_change:
        return 'Greater than average'
    if pct_change < mean_pct_change:
        return 'Less than everage'
    else:
        return 'Exactly average'
    
    outcome = change_checker(2)
    print(outcome)
```

```python
df['Change reatio'] = df['Percentage change (monthly) All property types'].apply(change_checker)
df
```

```python
mean_price = df['Average price All property types'].mean()
print(mean_price)
```

```python
def price_above_mean_checker(x):
    if x > mean_price:
        return 'Greater price than average'
    if x < mean_price:
        return 'Lower price than average'
    else:
        'Exactly the same as average'

df['Price above average?'] = df['Average price All property types'].apply(price_above_mean_checker)
df
```
