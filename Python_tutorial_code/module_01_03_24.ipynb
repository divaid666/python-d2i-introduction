```python
import pandas as pd

df = pd.DataFrame({'ChildId':['id1', 'id2', 'id3', 'id4', 'id5'],
                   'Age first contact':[6,12,11,1,19],
                   'Gender':['M','m', 'F', '', 'F' ],
                   'Birthday':['01/01/2002', '02/02/2003', pd.NA, '03/03/2023', '06/01/2012'],
                   'CP Plan?':['Y', 'n', 'N', 'No', 'yES'],})
print(df)
```

```python
df.info()
```

```python
#convert Birthday column to datetime
# when we apply something in pandas we do it row by row (row context)
df['Birthday'] = pd.to_datetime(df['Birthday'],format='%d/%m/%Y',errors='coerce') #erros - will handle NA values, without that it wouldn't work
print(df.info())
```

```python
df['Age at March 31'] = pd.to_datetime('31/03/2024',dayfirst=True) - df['Birthday']
print(df)
```

```python
df['Age at March 31'] = df['Age at March 31'] / pd.Timedelta('365 days')

df['Age at March 31'] = df['Age at March 31'].round().astype('int',errors='ignore')

print(df)
```

```python
df['Age as of today - 01.03.24'] = pd.to_datetime('01/03/2024',dayfirst=True)-df['Birthday']
df['Age as of today - 01.03.24'] = df['Age as of today - 01.03.24'] / pd.Timedelta('365 days')
df['Age as of today - 01.03.24'] = df['Age as of today - 01.03.24'].round().astype('int',errors='ignore')
df
```

```python
unique_genders = print(df['Gender'].unique())
print(unique_genders)
```

```python
gender_dict = {'M':'m',
               'm':'m',
               'F':'f',
               'f':'f'}
df['Gender'] = df['Gender'].map(gender_dict)

df
```

```python
df['Gender'] = df['Gender'].str.upper()
df
```

```python
df['Gender'] = df['Gender'].fillna('Not recorded')
df
```

```python
cp_cleaner = lambda row: 'y' if 'y' in row else 'n' if 'n' in row else pd.NA

df['CP Plan?'] = df['CP Plan?'].str.lower().apply(cp_cleaner) # better than a loop

print(df)
```
